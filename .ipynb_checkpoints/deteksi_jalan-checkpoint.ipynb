```python
from ultralytics import YOLO

# Load model dasar (YOLOv8n = ringan, atau YOLOv8s/m/l/x)
model = YOLO('yolov8n.pt')  # atau 'yolov8s.pt' untuk performa lebih tinggi

# Training
model.train(
    data='datasets/data.yaml',  # GANTI sesuai lokasi file yaml
    epochs=2,
    imgsz=640,
    batch=8,
    name='yolov8_vehicle_detect',
    patience=20,
    workers=4,
)
```

```python
import cv2
from ultralytics import YOLO
from deep_sort_realtime.deepsort_tracker import DeepSort

# Load model hasil training custom dari dataset Pothole
model = YOLO("best.pt")  # Ganti dengan model hasil training kamu dari Roboflow

# Load video uji coba
video_path = "testing3.mp4"  # Ganti dengan video uji coba kamu
cap = cv2.VideoCapture(video_path)

# Siapkan output video
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
out = cv2.VideoWriter("pothole_tracking_output.mp4", cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

# Inisialisasi Deep SORT
tracker = DeepSort(max_age=30)

# Set untuk menyimpan ID pothole unik yang pernah muncul
tracked_ids = set()

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Deteksi pothole dengan YOLOv8
    results = model.predict(frame, conf=0.4, verbose=False)[0]

    detections = []
    for box in results.boxes:
        x1, y1, x2, y2 = box.xyxy[0]
        conf = float(box.conf[0])
        cls = int(box.cls[0])  # Akan selalu 0 (karena hanya 1 class: pothole)

        # Format (x, y, w, h)
        detections.append(([x1, y1, x2 - x1, y2 - y1], conf, cls))

    # Update tracker
    tracks = tracker.update_tracks(detections, frame=frame)

    for track in tracks:
        if not track.is_confirmed():
            continue

        track_id = track.track_id
        ltrb = track.to_ltrb()
        x1, y1, x2, y2 = map(int, ltrb)

        # Tampilkan bounding box dan ID
        label = f"Pothole ID:{track_id}"
        cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
        cv2.putText(frame, label, (x1, y1 - 5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)

        tracked_ids.add(track_id)

    # Tampilkan jumlah total pothole unik yang pernah muncul
    cv2.putText(frame, f"Total Unique Potholes: {len(tracked_ids)}", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

    out.write(frame)
    cv2.imshow("Pothole Tracking", frame)
    if cv2.waitKey(1) == 27:  # Tekan ESC untuk keluar
        break

# Cleanup
cap.release()
out.release()
cv2.destroyAllWindows()
```
